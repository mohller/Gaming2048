# Simulating a batch of games

```python
mvals, maxval = [], []
for _ in range(50000):
    initial, moves, pops = generate_game(4)
    mvals.append(len(moves))
    maxval.append(2**np.max(playback(initial, moves, pops)[0]))
```

```python
for mxval in sorted(set(maxval)):
    numbers = [nv for nv, mx in zip(mvals, maxval) if mx == mxval]

    if numbers != []:
        plt.hist(numbers, bins=np.linspace(10, max(maxval), 50), alpha=.4, label=int(mxval))

plt.legend(title='Maximum')
plt.xlabel('Number of moves')
plt.ylabel('Frequency')
```

```python
percentage = lambda x: f'{x:2.0f}%' if x > 1 else ''
plt.pie([maxval.count(mx) for mx in sorted(set(maxval))], autopct=percentage, pctdistance=1.2);
```
